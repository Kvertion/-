```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Анализ данных индийских авиакомпаний

flight: Код рейса

```python
df = pd.read_csv('airlines_flights_data.csv')
df.rename(columns={'index':'id'},inplace=True)
df.set_index('id',inplace=True)
df.head()
```

Строю гистограмму и boxplot цен на билет
На гистограмме видно что большинство билетов сосредоточено в диапазоне до 20_000 при этом по мере увеличения цены количество билетов 
уменьшается и билеты более 100_000 практически не продаются, количество билетов логарифмировано чтобы было видно все значения

Boxplot наглядно показывает разброс цен, половина билетов находится в диапазоне от 5_000 до 45_000 также видны редкие выбросы которые показывают
значения больше 100_000

```python
fig,ax=plt.subplots(1,2,figsize=(15,6))
all_prices=df['price']
ax[0].set(ylabel='log количество',xlabel='цена',title='Гистограмма цен на билеты')
ax[0].hist(all_prices,log=True,edgecolor='black',label='asda');
ax[1].set(ylabel='количество',xlabel='boxplot цен на билеты');
ax[1].boxplot(all_prices);
# ax[1].set?
```

```python
Считаю среднюю цену билета для каждого города и строю гистограммы и боксплоты для каждого
```

```python
df.groupby(['source_city'])['price'].mean()#
```

```python
cities=df.groupby(['source_city'])['price'].mean()
cities=cities.index
```

```python
all_prices=df[['price','source_city']]
fig,ax=plt.subplots(4,3,figsize=(20,20))
fig.suptitle('Цена полета в каждый город')
fig.subplots_adjust(wspace=0.4,hspace=0.3)
# ax[0][1].hist(all_prices,log=True,bins=20)
for i in range(2):
    for j in range(3):
        city=all_prices[all_prices['source_city']==cities[i*3+j]]
        ax[i][j].set(title=f'prices to {cities[i*3+j]}')
        ax[i][j].hist(city['price'],log=True,edgecolor='black')
        ax[2+i][j].set(title=f'prices to {cities[i*3+j]}')
        ax[2+i][j].boxplot(city['price'])
```

Как видно гистограммы довольно схожи что говорит о схожести цен во все представленных города

Боксплоты также похожи, медианные и квартильные значения у всех выборок примерно одинаковы кроме Дели. Медианная цена в дели меньше на 2_000 и верхнаяя квартиль меньше чем у остальных городов, верхний ус заметно ниже и больше выбросов. Все это говорит о том что цены на полеты в Дели в среднем меньше чем в другие города

Смотрю самые дешевые и самые дорогие значения в выборке

Строю heatplot средней цены на билет по городам и авикомпаниям, как видно из графика самыми дорогими компаниями являются AirAsia и Air_India. А самыми недорогими городами для посещения Худерабад, Дели и Мумбаи

```python
fig,ax=plt.subplots(figsize=(10,6))
pivot=df.pivot_table(values='price',index='airline',columns='source_city',aggfunc='mean')
ax.set(xticks=range(6),yticks=range(6),xticklabels=cities,yticklabels=df['airline'].unique())
ax_=ax.imshow(pivot.values,cmap='Blues');
fig.colorbar(ax_);
```

Строю боксплот для эконом и бизнес класса, как видно ниже в эконом классе данные крайне однородные, разница между верхней и нижней квартилью незначительна, максимальное и минимальное значения крайне близки к медиане. Все это говорит о том что цены сосредоточены в узком диапазоне а редкие выбросы лишь выбиваются из картины
В бизнес классе ситуация другая, тут большой размах данных, максимум и минимум находятся далеко друг от друга, верхняя и нижняя квартиль находятся дальше чем в экономе. все это указывает на широкий разброс цен

```python
fix,ax=plt.subplots(figsize=(10,6))
economy=df.loc[df['class']=='Economy','price']
business=df.loc[df['class']=='Business','price']
ax.boxplot([economy,business],tick_labels=['Economy','Business']);
```

Строю линейный график зависимости цены от даты покупки билета. На нем видно что чем больше дней до покупки билета тем дешевле билет, и наоброт чем позже брать билет на рейс тем дороже он обойдется. Падения графика при приближении к нулю может говорить о скидках которые делает компания желая распродать все или об отменах билетов в последние дни 

```python
average_price = df.groupby('days_left')['price'].mean()
plt.plot(average_price);
```

Строю боксплоты зависимости цен от количества пересадок. 
График когда нет пересадок и есть 2 или более пересадки практически идентичен, чуть более высокое положение второго можно обьяснить многочиленными выбросами. Это говорит о крайней однородности этих данных, цены на билет при таких условиях практически не меняются и находятся в одном узком диапазоне
При одной пересадке ситуация обратная, видно очень широкий разброс данных, однако нижняя квартиль близка к медиане и минимуму, верхняя же и максимум удалены от медиана. Это показывает что дешевые билеты при одной пересадке практически одинаковы, дорогие же имеют широкий разброс

```python
fix,ax=plt.subplots(figsize=(10,6))
dummies=pd.get_dummies(df['stops'])
all_stops=pd.concat([df,dummies],axis=1)
one_stop=df.loc[all_stops['one'],'price']
two_or_more_stops=df.loc[all_stops['two_or_more'],'price']
zero_stops=df.loc[all_stops['zero'],'price']
all_stops
ax.set(title='Зависимость цены от количества пересадок')
ax.boxplot([zero_stops,one_stop,two_or_more_stops],tick_labels=['zero','one','two or more']);
```
